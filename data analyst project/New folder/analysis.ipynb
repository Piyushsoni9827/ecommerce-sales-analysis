# E-commerce Sales Data Analysis

This project analyzes a retail superstore dataset using Python to uncover sales trends, regional performance, and profitability insights.

## Tools Used
- Python (Pandas, Matplotlib, Seaborn)
- Jupyter Notebook
- Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/superstore.csv.csv", encoding="latin1")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df["Sales"].sum()
```

```python
df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
```

```python
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales by Region")
plt.show()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Month"] = df["Order Date"].dt.month

df.groupby("Month")["Sales"].sum().plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
sns.scatterplot(x="Discount", y="Profit", data=df)
plt.title("Discount vs Profit")
plt.show()
```

## Insights

- Technology category generated highest sales
- Western region has maximum contribution
- High discount leads to lower profit
- Sales increase during festive months
